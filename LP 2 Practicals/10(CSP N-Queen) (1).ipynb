```python
def is_safe(board, row, col):
    # Check if it's safe to place a queen at board[row][col]
    n = len(board)
    
    # Check the column
    for i in range(row):
        if board[i][col] == 1:
            return False
    
    # Check upper left diagonal
    for i, j in zip(range(row-1, -1, -1), range(col-1, -1, -1)):
        if board[i][j] == 1:
            return False
    
    # Check upper right diagonal
    for i, j in zip(range(row-1, -1, -1), range(col+1, n)):
        if board[i][j] == 1:
            return False
    
    return True

def solve_n_queens(n):
    def backtrack(row):
        nonlocal solutions
        if row == n:
            solutions.append(["".join("Q" if cell == 1 else "." for cell in row) for row in board])
            return
        
        for col in range(n):
            if is_safe(board, row, col):
                board[row][col] = 1
                backtrack(row + 1)
                board[row][col] = 0

    board = [[0] * n for _ in range(n)]
    solutions = []
    backtrack(0)
    return solutions

if __name__ == "__main__":
    try:
        # Input the board size (n)
        n = int(input("Enter the board size (n): "))

        # Solve the N-Queens problem for the specified board size
        solutions = solve_n_queens(n)

        # Print the solutions found
        if solutions:
            print(f"Found {len(solutions)} solution(s) for N-Queens problem:")
            for i, solution in enumerate(solutions):
                print(f"Solution {i + 1}:")
                for row in solution:
                    print(row)
                print()  # Add a blank line between solutions
        else:
            print("No solution found for N-Queens problem.")

    except ValueError:
        print("Invalid input. Please enter a valid integer for board size.")
```

```python
def is_safe(arr, x, y, n):
    # Check if it's safe to place a queen at arr[x][y] use this code
    
    # Check column attack
    for row in range(x):
        if arr[row][y] == 1:
            return False
    
    # Check upper left diagonal
    row, col = x, y
    while row >= 0 and col >= 0:
        if arr[row][col] == 1:
            return False
        row -= 1
        col -= 1
    
    # Check upper right diagonal
    row, col = x, y
    while row >= 0 and col < n:
        if arr[row][col] == 1:
            return False
        row -= 1
        col += 1
    
    return True

def nQueen(arr, x, n):
    # Solve N-Queens problem using backtracking
    
    if x >= n:
        return True
    
    for col in range(n):
        if is_safe(arr, x, col, n):
            arr[x][col] = 1
            if nQueen(arr, x + 1, n):
                return True
            arr[x][col] = 0
    
    return False

def main():
    n = int(input("Enter number of Queens: "))
    arr = [[0] * n for _ in range(n)]
    
    if nQueen(arr, 0, n):
        print("Found solution for N-Queens:")
        for i in range(n):
            for j in range(n):
                print(arr[i][j], end=" ")
            print()
    else:
        print("No solution found for N-Queens with", n, "Queens.")

if __name__ == '__main__':
    main()
```
